```python
data = {
 "intents": [
  {
   "tag": "greeting",
   "patterns": ["Hi","Hello","Good morning"],
   "responses": ["Hello! How can I help you?","Hi there!"]
  },
  {
   "tag": "password_reset",
   "patterns": ["reset password","forgot password"],
   "responses": ["Click reset password in settings"]
  },
  {
   "tag": "working_hours",
   "patterns": ["working hours","support time"],
   "responses": ["Support is available from 9AM to 6PM"]
  }
 ]
}
```

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.linear_model import LogisticRegression

patterns = []
tags = []

for intent in data["intents"]:
    for pattern in intent["patterns"]:
        patterns.append(pattern)
        tags.append(intent["tag"])

vectorizer = CountVectorizer()
X = vectorizer.fit_transform(patterns)

model = LogisticRegression()
model.fit(X,tags)
```

```python
import random

def chatbot(msg):
    X = vectorizer.transform([msg])
    tag = model.predict(X)[0]

    for intent in data["intents"]:
        if intent["tag"] == tag:
            return random.choice(intent["responses"])

print(chatbot("hello"))
print(chatbot("reset password"))
```

```python
import pickle

pickle.dump(model, open("model.pkl","wb"))
pickle.dump(vectorizer, open("vectorizer.pkl","wb"))
print("model.pkl and vectorizer.pkl saved successfully")
```
